```python
# Abschlussprojekt Data Science von Cheyco Marioni und Alexander Kenel

# Hypothese und Ablauf von unserem Data Science Abschlussprojekt:

# In der NZZ haben wir vor ein paar Monaten gelesen, dass der britische Premierminister Boris Johnson an Beliebtheit 
# gewonnen hat. Der Grund daf√ºr sind die Impferfolge Grossbritanniens im Kampf gegen Covid-19.
# Mit Hilfe von Twitter Tweets aus den St√§dten Liverpool und Birmingham wollten wir √ºberpr√ºfen, ob dies in Twitter
# ebenfalls so abgebildet ist. √úber den Zeitraum von Oktober 2020 bis Mai 2021 haben wir aus beiden St√§dten f√ºr jeden 
# Monat bis zu 500 Tweets gescrapt. Anschliessend haben wir die Tweets mit einer Sentiment-Analyse in positive und 
# negative Tweets unterteilt. Die Resultate haben wir mit einem dynamischen Balkendiagramm visualisiert. Die St√§dte 
# Liverpool und Birmingham haben wir bewusst ausgesucht. Liverpool hat wohl eher ein Image als Arbeiterstadt, w√§hrend
# Birmingham als wohlhabend gilt. Wir wollten sehen, ob das Verh√§ltnis von netativen zu positiven Tweets zu Boris
# Johnson je nach Stadt variert. 
```

```python
# Importieren der n√∂tigen Libraries. Beide mussten zuvor installiert werden f√ºr das installieren von Twitter Twint 
# hat uns der folgende Link sehr geholfen: https://pypi.org/project/twint/. 

# Twint ist f√ºr das Scrapen der Tweets auf Twitter zust√§ndig. 
# Mit c = twint.Config() konfigurieren wir die Suchparameter in der n√§chsten Zelle

import nest_asyncio
nest_asyncio.apply()

import twint
c = twint.Config()
```

```python
# Konfiguration der Suchkriterien

c.Search = "Boris Johnson"
c.Lang = "en"
c.Pandas = True
c.Limit = 500
c.Since = '2021-05-01'
c.Until = '2021-05-31'
c.Near = "Birmingham"
c.Store_csv = True
c.Output = "boris_johnson_birmingham_05.2021.csv"
twint.run.Search(c)
```

```python
def column_names():
    return twint.output.panda.Tweets_df.columns
def twint_to_pd(columns):
    return twint.output.panda.Tweets_df[columns]
```

```python
#anzeigen aller verf√ºgbaren Spalten/Daten

column_names()
```

```python
# Aussuchen der f√ºr uns relevanten Spalten mit Pandas

tweet_df = twint_to_pd(["date", "username", "tweet", "hashtags", "nlikes"])
```

```python
# anzeigen der ersten 10 Tweets mit den gew√§hlten Spalteninfos

tweet_df.head(10)
```

```python
print(tweet_df)
```

```python
tweet_list = tweet_df['tweet'].tolist()
```

```python
# Import der Library f√ºr die Sentiment-Analyse

from transformers import pipeline
```

```python
# Ausf√ºhrung der Sentiment-Analyse --> Einteilung in Positiv und Negativ mit einem Accuracy-Score

sentiment_classifier = pipeline('sentiment-analysis')
```

```python
# Speichern der Resultate der Sentiment-Analyse in einer Liste results

results = sentiment_classifier(tweet_list)
```

```python
# Darstellung der Tweets mit dem Label Positiv oder Negativ, sowie dem Score

ind = 0
for result in results:
    print('===========')
    print(tweet_list[ind]+'\n')
    print(f"label: {result['label']}, with score: {round(result['score'], 4)}")
    print('===========\n')
    ind = ind + 1
```

```python
# Darstellung aller Resultate nach Monat und Stadt, sowie positive und negative Tweets

import pandas as pd
import plotly.express as px 


df_L1_positiv = pd.read_csv("L1Pos.csv")
df_L1_negativ = pd.read_csv("L1Neg.csv")
df_L2_positiv = pd.read_csv("L2Pos.csv")
df_L2_negativ = pd.read_csv("L2Neg.csv")

def prepare(dfs, datefrom):
    result=[]
    for item in dfs:
        item.columns = ['month','Tweets', 'Category']
        # item = item[item['month']>= datefrom]
        result.append(item)
    return result 
dfs = [df_L1_positiv, df_L1_negativ, df_L2_positiv, df_L2_negativ]
categories = ['Liverpool Positiv', 'Liverpool Negativ', 'Birmingham Positiv', 'Birmingham Negativ'] 
i=0
while i < len(categories):
    dfs[i].loc[:,'category'] = categories[i]
    i = i+1     
df=prepare(dfs, datefrom)    
df = pd.concat(df, ignore_index=True)
display(df)
```

```python
# Positive und negative Tweets gesamt von Oktober 2020 bis Mai 2021 nach Stadt

fig = px.bar(
    df, x="Category", 
    y="Tweets",
    color="Category",
    color_discrete_map={
        'Birmingham Positiv': 'green',
        'Birmingham Negativ': 'red',
        'Liverpool Positiv': 'green',
        'Liverpool Negativ': 'red',
    }   
)
fig.update_layout(
    height=600,
    title_text='Total Tweets about Boris Johnson From October 2020 to May 2021'
)
fig.show()
```

```python
# Dynamisches Balkendiagramm nach Monat mit den Resultaten

fig = px.bar(
    df, 
    x="Category", 
    y="Tweets",
    animation_frame="month",
    animation_group="Category",
    range_y=[0,500],
    color="Category",
    color_discrete_map={
        'Birmingham Positiv': 'green',
        'Birmingham Negativ': 'red',
        'Liverpool Positiv': 'green',
        'Liverpool Negativ': 'red',
    }
)
fig.update_layout(
height =600,
title_text='Tweets about Boris Johnson From October 2020 to May 2021'
)
fig.show()
```

```python
# Darstellung des Verh√§ltnis von positiven zu negativen Tweets

import pandas as pd
import plotly.express as px 


df_L1_Ratio = pd.read_csv("L1Ratio.csv")
df_L2_Ratio = pd.read_csv("L2Ratio.csv")


def prepare(dfs, datefrom):
    result=[]
    for item in dfs:
        item.columns = ['month','Ratio P to N', 'Category']
        # item = item[item['month']>= datefrom]
        result.append(item)
    return result 
dfs = [df_L1_Ratio, df_L2_Ratio]
categories = ['Liverpool Ratio', 'Birmingham Ratio'] 
i=0
while i < len(categories):
    dfs[i].loc[:,'category'] = categories[i]
    i = i+1     
df=prepare(dfs, datefrom)    
df = pd.concat(df, ignore_index=True)
display(df)
```

```python
# Verh√§ltnis von positiven zu negativen Tweets nach Monat

fig = px.bar(
    df, 
    x="Category", 
    y="Ratio P to N",
    animation_frame="month",
    animation_group="Category",
    range_y=[0,1],
    color="Category",
    color_discrete_map={
        'Liverpool Ratio': 'blue',
        'Birmingham Ratio': 'grey',
    }
)
fig.update_layout(
height =600,
title_text='Ratio Positive to Negative Tweets'
)
fig.show()
```
